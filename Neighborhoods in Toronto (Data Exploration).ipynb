# Neighborhoods in Toronto
(Data Exploration)

### Import libraries

```python
import os
import pandas as pd
import folium as folium
```

### Read csv to dataframe

```python
n_loc = pd.read_csv('neighborhood_location.csv')
n_loc = n_loc.drop(columns=['Unnamed: 0'])
```

```python
n_loc
```

### Explore map from dataframe

```python
map = folium.Map(location=[43.806686, -79.194353], zoom_start=10)

for point in range(len(n_loc)):
    folium.Marker([n_loc['Latitude'][point], n_loc['Longitude'][point]], popup=n_loc['Postal Code'][point]).add_to(map)

map
```
